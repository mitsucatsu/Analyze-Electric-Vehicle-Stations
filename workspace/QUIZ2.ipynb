## Task 8

Electric vehicles have become more and more crucial to plans around climate change. Now, let's take a closer looka t stations that anyone can access.

Create a boolean mask that is `True` for any row of `stations` where `access` is public.

## Task 9

Create a boolean mask that is `True` for any row of `stations` where `fuel` is `electric`.

```python
electric_fuel_mask = stations['fuel'] == 'electric'
electric_fuel_stations = stations[electric_fuel_mask]
electric_fuel_stations.head()
```

## Task 10

Use the boolean masks from **Task 8** and **Task 9** to filter `stations` down to only rows that are both public-access and electric-fuel.

# Task 18

The maximum number of stations is quite a bit larger than the 75th percentile. Let's look at the rows between these values.

Create a boolean mask that is `True` in each row of `not_privately_owned` where `number_of_stations` is bigger than `17` or the 75th percentile.

```python
percentile_75 = not_privately_owned['number_of_stations'].quantile(0.75)

mask = not_privately_owned['number_of_stations'] > max(17, percentile_75)

filtered_not_privately_owned = not_privately_owned[mask]
filtered_not_privately_owned
```
